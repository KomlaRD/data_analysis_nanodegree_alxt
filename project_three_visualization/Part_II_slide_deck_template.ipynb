## Investigation Overview

This investigation explored factors associated with diabetes among females of Pima Indian Heritage 21 years and older.
Participants with diabetes were on average older than participants without diabetes. The chances of diabetes increased with increasing number of pregnancies, blood pressure, triceps skinfold thinckness and BMI. Participants with diabetes had higher amounts of serum insulin compared to participants without diabetes and a higher Diabetes Pedigree Funtion on average was also associated with increased chances of having diabetes. Participants with diabetes had higher blood glucose 2 hours after oral glucose test.

## Dataset Overview

The dataset consisted of 768 observation with 9 features. The main outcome in the dataset was diabetes among females of Pima Indian Heritage. Explanatory features in the dataset includes number of pregnancies, serum glucose, insulin, blood pressure, tricep skinfold thickness, BMI, Diabetes Pedigree Function, and age.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
diabetes_plot = pd.read_csv("data/diabetes_explanatory.csv")
```

## (Visualization 1)

Number of pregnancies is associated with the outcome of diabetes among females of Pima Indian heritage. The average number of pregnancies for participants without diabetes is 2 where as that of participants with diabetes is 3.

```python
# Relationship between number of pregnancies and diabetes
# Box plot
plt.figure(figsize=[10,5])
sns.boxplot(data = diabetes_plot, x = "Outcome", y = "Pregnancies");
plt.title("Number of pregnacies are higher in participants with diabetes vs participants without diabetes");
```

## (Visualization 2)

The average pressure of participants with diabetes is higher compared to participants without diabetes. However, participants without diabetes have overall higher blood pressure compared to participants without diabetes.

```python
# Relationship between blood pressure and diabetes
# Box plot
plt.figure(figsize=[10,5])
sns.boxplot(data = diabetes_plot, x = "Outcome", y = "BloodPressure");
plt.title("Average blood pressure is higher in participants with diabetes vs participants without diabetes");
```

## (Visualization 3)

Average skinfold thickness in participants with diabetes is higher compared to that of participants without diabetes. However, participants without diabetes have higher skinfold thickness overall compared to participants without diabtes.

```python
# Relationship between triceps skinfold thickness and diabetes
# Box plot
plt.figure(figsize=[10,5])
sns.boxplot(data = diabetes_plot, x = "Outcome", y = "SkinThickness");
plt.title("Skinfold thickness in participants with diabetes is higher on average compared to participants without diabetes");
```

## (Visualization 4)

Average insulin concentration in participants with diabetes is higher compared to that of participants without diabetes. 

```python
# Relationship between insulin and diabetes
# Box plot
plt.figure(figsize=[10,5])
sns.boxplot(data = diabetes_plot, x = "Outcome", y = "Insulin");
plt.title("Insulin concentration is higher in participants with diabetes vs participants without diabetes");
```

## (Visualization 5)

Average body mass index (BMI) in participants with diabetes is higher compared to that of participants without diabetes. However, participants without diabetes have higher BMI overall compared to participants without diabtes

```python
# Relationship between BMI and diabetes
# Box plot
plt.figure(figsize=[10,5])
sns.boxplot(data = diabetes_plot, x = "Outcome", y = "BMI");
plt.title("Average BMI is higher in participants with diabetes vs participants without diabetes");
```

## (Visualization 6)

Diabetes pedigree function is on average higher in participants with diabetes is higher compared to that of participants without diabetes.

```python
# Relationship between diabetes pedigree function and diabetes
# Box plot
plt.figure(figsize=[10,5])
sns.boxplot(data = diabetes_plot, x = "Outcome", y = "DiabetesPedigreeFunction");
plt.title("Average diabetes pedigree function is higher in participants with diabetes vs participants without diabetes");
```

## (Visualization 7)

Serum glucose after 2h oral glucose tolerance test in participants with diabetes is higher compared to that of participants without diabetes. 

```python
# Relationship between serum glucose and diabetes
# Box plot
plt.figure(figsize=[10,5])
sns.boxplot(data = diabetes_plot, x = "Outcome", y = "Glucose");
plt.title("Serum glucose is higher in participants with diabetes vs participants without diabetes");
```

## (Visualization 8)

Participants with diabetes are on average older compared to participants without diabetes.

```python
# Relationship between age and diabetes
# Box plot
plt.figure(figsize=[10,5])
sns.boxplot(data = diabetes_plot, x = "Outcome", y = "Age");
plt.title("Participants with diabetes are older compared to participants without diabtes");
```

## (Visualization 9)

Participants with diabetes have higher concentrations of both glucose and insulin.

```python
# Relationship between glucose and insulin stratified by outcome of diabetes
# Scatter plot with color encoding
plt.figure(figsize=[10,5])
plt.scatter(data = diabetes_plot, x = "Insulin", y = "Glucose", c = "Outcome", cmap = "vlag")
plt.title("Participants with diabetes have higher glucose and insulin levels");
plt.colorbar(ticks = [0,1]);
```
